```python
# Assuming fav_movie contains the selected movie's data

fav_movie_embedding = embedding  # This should be the embedding for the selected movie

# Ensure the embedding is correctly retrieved and unique
print(f"Embedding for selected movie: {fav_movie['title']}")
print(fav_movie_embedding)

# Search for similar movies in the FAISS index
D, I = index.search(fav_movie_embedding, 5)

# Debug: Check distances and indices
print("Distances: ", D)
print("Indices: ", I)

# Fetch and print the top matches
best_matches = np.array(cleaned_data['String_data'])[I.flatten()]
for i in best_matches:
    print("Next movie to watch:")
    print(i)
    print()
```

```python
import numpy as np

def combine_user_embeddings(user_history):
    embeddings = [np.array(entry['embedding']) for entry in user_history['watch_history']]
    
    num_movies = len(embeddings)
    
    # Generate weights: more recent movies have higher weights
    weights = np.linspace(1, num_movies, num_movies)[::-1]
    
    # Normalize the weights so they sum to 1
    weights /= weights.sum()
    
    # Debugging: Print embeddings and weights
    print("Embeddings:", embeddings)
    print("Weights:", weights)
    
    combined_embedding = np.average(embeddings, axis=0, weights=weights)
    
    # Debugging: Print combined embedding
    print("Combined Embedding:", combined_embedding)
    
    return combined_embedding

user_combined_embedding = combine_user_embeddings(user_watch_history)

# Debugging: Print reshaped embedding
print("Reshaped Combined Embedding for FAISS:", user_combined_embedding.reshape(1, -1))

# Use this combined embedding to get recommendations
D, I = index.search(user_combined_embedding.reshape(1, -1), 5)
best_matches = np.array(cleaned_data['String_data'])[I.flatten()]

# Print recommendations
for i in best_matches:
    print("Next movie to watch:")
    print(i)
    print()
```

```python
import json

# Assuming `best_matches` contains the movie strings as shown
recommendations = []
for i in best_matches:
    movie = {}
    lines = i.split('\n')
    
    for line in lines:
        line = line.strip() 
        if line.startswith("Type:"):
            movie["Type"] = line.replace("Type:", "").strip()
        elif line.startswith("Title:"):
            movie["Title"] = line.replace("Title:", "").strip().rstrip(',')
        elif line.startswith("Director:"):
            movie["Director"] = line.replace("Director:", "").strip().rstrip(',')
        elif line.startswith("Cast:"):
            movie["Cast"] = [actor.strip() for actor in line.replace("Cast:", "").strip().rstrip(',').split(",")]
        elif line.startswith("Released:"):
            movie["Released"] = int(line.replace("Released:", "").strip().rstrip(','))
        elif line.startswith("Rating:"):
            movie["Rating"] = line.replace("Rating:", "").strip().rstrip(',')
        elif line.startswith("Genre:"):
            movie["Genre"] = [genre.strip() for genre in line.replace("Genre:", "").strip().rstrip(',').split(",")]
        elif line.startswith("Description:"):
            movie["Description"] = line.replace("Description:", "").strip()

    recommendations.append(movie)

# Convert to JSON
json_output = json.dumps(recommendations, indent=4)

# Save to file
with open('recommendations.json', 'w') as file:
    file.write(json_output)

print(json_output)
```
